```python
import matplotlib.pyplot as plt


def plot_closing_prices(crypto_name):
    crypto = data[data['Name'] == crypto_name]
    
    plt.figure(figsize=(10, 6))
    plt.plot(crypto['Date'], crypto['Close'], label=f'{crypto_name} Closing Price', color='blue')
    plt.title(f'Closing Prices Over Time for {crypto_name}')
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.grid(True)
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Example: Plotting for Aave
plot_closing_prices('Aave')
```

```python


# Filter the data for a specific cryptocurrency, e.g., 'Bitcoin'
crypto = data[data['Name'] == 'Bitcoin']

# Plot the closing prices over time
plt.figure(figsize=(10, 6))
plt.plot(crypto['Date'], crypto['Close'], label='Closing Price', color='blue')
plt.title('Closing Prices Over Time for Bitcoin')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python

max_marketcap = data.groupby('Name')['Marketcap'].max().reset_index()

# Sort the cryptocurrencies by their maximum market capitalization
max_marketcap_sorted = max_marketcap.sort_values(by='Marketcap', ascending=False).head(10)

# Plot the top 10 cryptocurrencies by maximum market capitalization
plt.figure(figsize=(10, 6))
plt.barh(max_marketcap_sorted['Name'], max_marketcap_sorted['Marketcap'], color='green')
plt.title('Top 10 Cryptocurrencies by Maximum Market Capitalization')
plt.xlabel('Market Capitalization (USD)')
plt.ylabel('Cryptocurrency')
plt.tight_layout()

# Show the plot
plt.show()

# Display the cryptocurrency with the highest market capitalization
print("Cryptocurrency with highest market capitalization:", max_marketcap_sorted.iloc[0])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("C:\\Users\\HIMANSHU\\Downloads\\Crypto project\\cleaned_sorted_output.csv")

# Check for null values and drop rows with missing data (if any)
data_cleaned = data.dropna(subset=['Volume', 'Marketcap'])

# Calculate the correlation
correlation = data_cleaned['Volume'].corr(data_cleaned['Marketcap'])
print(f'Correlation between Volume and Market Capitalization: {correlation}')

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Volume', y='Marketcap', data=data_cleaned, color='purple')
plt.title('Volume Traded vs. Market Capitalization')
plt.xlabel('Volume Traded')
plt.ylabel('Market Capitalization')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# 3. Highest and lowest prices for top cryptocurrency
crypto = data[data['Name'] == 'Ethereum']  # Example: Ethereum
plt.figure(figsize=(10, 6))
plt.plot(crypto['Date'], crypto['High'], label='High Price', color='green')
plt.plot(crypto['Date'], crypto['Low'], label='Low Price', color='red')
plt.title('High and Low Prices for Ethereum Over Time')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# 4. Average daily price range for top cryptocurrencies
data['Price Range'] = data['High'] - data['Low']
avg_price_range = data.groupby('Name')['Price Range'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(avg_price_range.index, avg_price_range.values, color='purple')
plt.title('Average Daily Price Range for Top Cryptocurrencies')
plt.xlabel('Cryptocurrency')
plt.ylabel('Average Price Range (USD)')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. Correlation between volume traded and market capitalization
plt.figure(figsize=(10, 6))
plt.scatter(data['Volume'], data['Marketcap'], color='blue', alpha=0.5)
plt.title('Correlation between Volume Traded and Market Capitalization')
plt.xlabel('Volume Traded')
plt.ylabel('Market Capitalization (USD)')
plt.grid(True)
plt.show()

# Continue with si
```
